```python
import requests
import pandas as pd
import folium
import ast
```

```python
url="https://zoom.earth/data/storms/?id=beryl-2024&lang=es"
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
    
}
```

```python
r = requests.get(url, headers=headers)
r.status_code
```

```python
data=r.json()
data.keys()
```

```python
fecha=[]
viento=[]
presion=[]
descripcion=[]
coordenadas=[]

for i in range(len(data["track"])):
    fecha.append(data["track"][i]["date"])
    viento.append(data["track"][i]["wind"])
    presion.append(data["track"][i]["pressure"])
    descripcion.append(data["track"][i]["description"])
    coordenadas.append(data["track"][i]["coordinates"])
    
#Eliminar corchetes de la columna coordenadas
coordenadas = [str(i).replace("[","").replace("]","") for i in coordenadas]
```

```python
beryl=pd.DataFrame({"fecha":fecha,"viento":viento,"presion":presion,"descripcion":descripcion,"coordenadas":coordenadas})
#Transformar en latitud y longitud
beryl["longitud"]=beryl["coordenadas"].apply(lambda x: float(x.split(",")[0]))
beryl["latitud"]=beryl["coordenadas"].apply(lambda x: float(x.split(",")[1]))
beryl
```

```python
#Mapa
m=folium.Map(location=[beryl["latitud"].mean(),beryl["longitud"].mean()],zoom_start=5,tiles="https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}",attr="Google")
for i in range(len(beryl)):
    folium.CircleMarker(location=[beryl["latitud"][i],beryl["longitud"][i]],radius=5,color="orange",fill=True,
                        tooltip="<b>Tipo:</b> "+beryl["descripcion"][i]+"<br><b>Fecha:</b> "+beryl["fecha"][i]+"<br><b>Viento:</b> "+str(beryl["viento"][i])+" km/h<br>").add_to(m)
    #Dibujar línea 
    if i>0:
        folium.PolyLine(locations=[[beryl["latitud"][i-1],beryl["longitud"][i-1]],[beryl["latitud"][i],beryl["longitud"][i]]],color="red").add_to(m)
#Salvar mapa
m.save("beryl.html")    
m
```
